# Part D - ForkTheChains

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output_forkthechain.csv', delimiter=',', header=None, names=['Fork Name', 'Year-Month-Day', 'Gas Price'])
df

df['Fork Name'] = df['Fork Name'].str.replace("(","")
df['Fork Name'] = df['Fork Name'].str.replace("'","")
df['Year-Month-Day'] = df['Year-Month-Day'].str.replace("'","")
df['Year-Month-Day'] = df['Year-Month-Day'].str.replace(")","")
df['Gas Price'] = df['Gas Price'].str.replace(")","")
df
```

```python
for index, row in df.iterrows():
  if len(row['Year-Month-Day'].split('-')[1]) == 1:
    df.loc[index, 'Year-Month-Day'] =  row['Year-Month-Day'].split('-')[0] + "-0" + row['Year-Month-Day'].split('-')[1] + "-" + row['Year-Month-Day'].split('-')[2]
  if len(row['Year-Month-Day'].split('-')[2]) == 1:
    df.loc[index, 'Year-Month-Day'] =  row['Year-Month-Day'].split('-')[0] + "-" + row['Year-Month-Day'].split('-')[1] + "-0" + row['Year-Month-Day'].split('-')[2]

df
```

## Constantinople

```python
df_constantinople = df[df['Fork Name'] == 'Constantinople']

df_constantinople = df_constantinople.sort_values(by=['Year-Month-Day'])
df_constantinople['Gas Price'] = pd.to_numeric(df_constantinople['Gas Price'])
df_constantinople.plot(kind='line', x='Year-Month-Day', y='Gas Price', figsize=(18,5), xlabel='Year-Month-Day', ylabel='Avg Gas Price', title='Gas Price vs Time for Fork - Constantinople')
```
